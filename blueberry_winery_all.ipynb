```python
wine_con = pd.concat([white_wine, red_wine], axis=0)
wine_con
```

```python
wine_con.reindex(columns=['type','residual_sugar','total_sulfur_dioxide','sulphates','alcohol','volatile_acidity','quality','quality_label'], level=1)
wine_con=wine_con[['type','residual_sugar','total_sulfur_dioxide','sulphates','alcohol','volatile_acidity','quality','quality_label']]
wine_con
```

```python
plt.pie(wine_con['type'].value_counts(dropna=False), labels=['Red', 'White'], colors = ['red','green'],startangle=90, autopct='%.0f%%') 
plt.legend(title='Wine Type')
```

```python
fig, axes = plt.subplots(nrows =1, ncols=2,figsize=(10,8))
sns.histplot(data=red_wine, x='quality', ax=axes[1], color='red').set_title("Red Wine")
sns.histplot(data=white_wine, x='quality', ax=axes[0], color='green').set_title("White Wine")
```

```python
fig, axes = plt.subplots(nrows =1, ncols=2,figsize=(10,8))
sns.boxplot(data = red_wine, x = 'quality_label', y = 'alcohol', ax=axes[1], color = 'red').set_title("Red Wine")
sns.boxplot(data = white_wine, x = 'quality_label', y = 'alcohol', ax=axes[0], color = 'green').set_title("White Wine")
```
