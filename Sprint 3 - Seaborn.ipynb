```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Distribution Plots

Displot


```python
sns.set_theme(style = 'whitegrid')
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 100)
```

### Jointplot


```python
#hex, kde, 
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

### Pairplot

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.kdeplot(tips['total_bill'])
```

### Categorical Data Plots

```python
### Bar plot

sns.barplot(data = tips, x = 'sex', y = 'total_bill', estimator = np.std)
```

### Count Plot

```python
sns.countplot(x='sex', data = tips, palette = 'rainbow')
```

### Box Plot

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(data = tips, palette = 'rainbow', orient = 'h')
```

```python
sns.boxplot(x='day', y='total_bill', data = tips)
```

### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data = tips, hue = 'sex', palette = 'Set1', split = True)
```

### Stripplot and Swarmplot

```python
sns.stripplot(x='day', y ='total_bill', data=tips,jitter=True, hue = 'sex', dodge = True)
```

```python
sns.swarmplot(x='day', y ='total_bill', data = tips)
```

```python
sns.violinplot(x='tip', y= 'day',data = tips, palette = 'rainbow')
sns.swarmplot(x='tip', y ='day', data = tips, color = 'black', size = 3)
```

### Factorplot

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

### Regression Plots


```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', palette = 'rainbow', markers=['o', 'v'], scatter_kws = {'s':100})
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col = 'sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex', aspect =1.6, height = 12)
```

### HeatMap

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot = True, cmap='coolwarm')
```
