```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfw = pd.read_csv(r'C:\Users\janin\OneDrive\Desktop\wine+quality\winequality-white.csv',sep=';') # Gathering data
dfw.head(5)
```

```python
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="quality", y="alcohol", data=dfw)
```

```python
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lmplot(x="quality", y="alcohol", data=dfw)
```

```python
sns.set_theme(style="whitegrid")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="quality",y = "fixed acidity",
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 8), linewidth=0,
                data=dfw, ax=ax)
```

```python
dfw.hist(bins=25,figsize=(10,10))
# display histogram
plt.show()
```
