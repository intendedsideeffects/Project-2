```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
# Functional Method 

plt.plot(x,y)

# add labels and title 
plt.xlabel('Time')
plt.ylabel('Space')
plt.title('Time and Space Relationship')

plt.show()
```

```python
# creating multi plots on the same canvas

plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
# Object oriented method

fig = plt.figure() # imagine majory blank campus 
axes = fig.add_axes([0.1,0.1, 0.8, 0.8])


axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```
